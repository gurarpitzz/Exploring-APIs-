```python
import requests

def fetch_github_user_profile(username):
    url = f"https://api.github.com/users/{username}"
    response = requests.get(url)

    if response.status_code == 200:
        user_data = response.json()
        return user_data
    else:
        print(f"Error: {response.status_code} - {response.json().get('message')}")
        return None

if __name__ == "__main__": # Changed _name_ to __name__
    username = input("Enter the GitHub username: ")
    profile = fetch_github_user_profile(username)

    if profile:
        print("User Profile:")
        print(f"Login: {profile['login']}")
        print(f"ID: {profile['id']}")
        print(f"Node ID: {profile['node_id']}")
        print(f"Avatar URL: {profile['avatar_url']}")
        print(f"HTML URL: {profile['html_url']}")
        print(f"Followers: {profile['followers']}")
        print(f"Following: {profile['following']}")
        print(f"Public Repos: {profile['public_repos']}")
        print(f"Bio: {profile.get('bio')}")
```
